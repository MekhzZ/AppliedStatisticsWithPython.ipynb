```python
# creating dataframe

l = [['x',10],['y',20],['z',30]]
df = pd.DataFrame(l,columns = ['name','age'])
df
```

```python
#using dictionaries

d = {'name':['x','y','z'],'age':[10,20,30]}
df = pd.DataFrame(d)
df
```

```python
# combining multiple lists into a dataframe

l1 = ['x','y','z']
l2 = [10,20,30]
c1 = list(zip(l1,l2))
df = pd.DataFrame(c1,columns = ['name','age'])
df
```
