```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime, timedelta

# Define the activities and their durations
activities = [
    ("applied statistics with Python", 10),
    ("IIC Hackathon", 4),
    ("Continued applied statistics with Python", 4),
    ("data mining", 5),
    ("hand sign detection project", 8),
    ("Continued data mining", 13),
    ("Tableau doughnut chart", 1),
    ("Birch algorithm", 1),
    ("NumPy", 2),
    ("Tableau dashboard", 1),
    ("data visualizer web app", 1),
    ("customer segmentation analysis", 1),
    ("Continued NumPy", 1),
    ("pandas", 2),
    ("Data wrangling", 2),
    ("Matplotlib and Seaborn", 1),
    ("data understanding and insights", 3)
]

# Calculate start and end dates for each activity
start_date = datetime(2024, 6, 1)
dates = [(start_date + timedelta(days=sum(d for _, d in activities[:i])), d) for i, (a, d) in enumerate(activities)]

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Add activities to the plot
for (activity, duration), (start, _) in zip(activities, dates):
    end = start + timedelta(days=duration)
    ax.plot([start, end], [activity, activity], marker='o')

# Format the dates on the x-axis
ax.xaxis.set_major_locator(mdates.WeekdayLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))

#title
plt.title('60DaysOfLearning2024 - LearningWithLeapfrog')
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
